```
%sql

create or replace temporary view view_notebooks as (
select 
  T.database as `database`,
  T.payload.id as id,
  T.payload.type as type, 
  from_json(cast(T.payload.data as string), "name STRING, size INT, reposExportFormat STRING, blobPath STRING, lastSyncedBlobPath STRING, contentSha256Hex STRING, language STRING, created_timestamp BIGINT, updated_timestamp BIGINT") as data
  from delta.`/mnt/indices/etl/webapp/tree/snapshot/delta` T
  where T.payload.type = 'shell'
)
```

```
%sql
select count(*) from view_notebooks where data.blobPath is not null and !(data.blobPath <=> data.lastSyncedBlobPath);
```
